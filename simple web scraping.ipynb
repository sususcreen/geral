```python
url = "https://www.imdb.com/list/ls520913000/?ref_=hm_edcft_c_ft_lst_csegmybo_mov_1_cta"
```

```python
html = requests.get(url)
s = BeautifulSoup(html.content, 'html.parser')
```

```python
print(s.prettify())
```

```python
scrapped_titles= s.find_all('h3', class_="lister-item-header")
```

```python
titles=[]
for title in scrapped_titles:
    title = title.get_text().replace('\n', '')
    title = title.strip(' ')
    titles.append(title)
    
titles
```

```python
scrapped_genre = s.find_all('span', class_="genre")

genres=[]
for genre in scrapped_genre:
    genre = genre.get_text().replace('\n', '')
    genre = genre.strip(' ')
    genres.append(genre)
    
genres
```

```python
scrapped_rate = s.find_all('div', class_="ipl-rating-star small")

ratings=[]
for rate in scrapped_rate:
    rate = rate.get_text().replace('\n', '')
    rate = rate.strip(' ')
    ratings.append(rate)
    
ratings

```

```python
scrapped_runtime= s.find_all('span', class_="runtime")

runtime = []

for time in scrapped_runtime:
    time = time.get_text().replace('\n', "")
    time = time.strip(" ")
    runtime.append(time)
    
runtime
```

```python
scrapped_rank= s.find_all('span', class_="lister-item-index unbold text-primary")

rank = []

for number in scrapped_rank:
    number= number.get_text().replace('\n', "")
    number= number.strip(" ")
    rank.append(number)
    
rank
```

```python
import pandas as pd
```

```python
scrapped_years= s.find_all('span', class_="lister-item-year text-muted unbold")

year = []

for years in scrapped_years:
    years= years.get_text().replace('\n', "")
    years= years.strip(" ")
    year.append(years)



data = {'Rank':rank, 'Title':titles, 'Genre':genres, 'Rating':ratings, 'Year':year}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Year'].replace(to_replace='(I) (2023)', value="(2023)", inplace=True)
```

```python
df['Year'].replace(to_replace='(2023)', value="2023", inplace=True)
df
```
